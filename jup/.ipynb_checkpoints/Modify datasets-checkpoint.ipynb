```python
import pandas
df = pandas.read_csv("supermarkets.csv")
```

```python
df
```

```python
# How can we convert into latitude and longtitude the whole table?
```

```python
df["Address"]=df["Address"]+", "+df["City"]+", "+df["State"]+", "+df["Country"]
```

```python
df
```

```python
df["Coordinates"]=df["Address"].apply(nom.geocode)
```

```python
df
```

```python
df.Coordinates[0]
```

```python
df.Coordinates[0].latitude
```

```python
df["Latitude"]=df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
```

```python
df
```

```python
df["Longtitude"]=df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
```

```python
df
```
